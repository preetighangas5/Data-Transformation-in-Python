**DATA TRANSFORMATION**

**Question 1:  Handle Missing Values** (Basic Cleaning)

Fill missing values using the following:

 - Age → Median

 - Salary → Mean

 - Hours_Worked_Per_Week → Median

 - Performance_Score → Mean

Display the dataset after handling missing values.

**Answer:**

```python
import pandas as pd
df = pd.read_csv('/content/employee_productivity_dataset.csv')

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score'] = df['Performance_Score'].fillna(df['Performance_Score'].mean())

print("Missing values after handling:\n", df.isnull().sum())
print("\nDataset preview after Q1:\n", df.head())
```

**Question 2:  Label Encoding**

Convert these columns into numeric values using Label Encoding:

 - Gender

 - Department

Show the updated columns.

**Answer:**

```python
from sklearn.preprocessing import LabelEncoder

le_gender = LabelEncoder()
le_dept = LabelEncoder()

df['Gender_Encoded'] = le_gender.fit_transform(df['Gender'])
df['Department_Encoded'] = le_dept.fit_transform(df['Department'])

print("Gender Mapping:", dict(zip(le_gender.classes_, range(len(le_gender.classes_)))))
print("Department Mapping:", dict(zip(le_dept.classes_, range(len(le_dept.classes_)))))
print(df[['Gender', 'Gender_Encoded', 'Department', 'Department_Encoded']].head())
```

**Question 7:  Build Preprocessing Pipeline**

Create a pipeline that:

 -  Applies encoding to categorical columns

 -  Applies scaling to numerical columns

Use:

 - ColumnTransformer

 - Pipeline

 **Answer:**

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Load the dataset (assuming it's already preprocessed from Q1-Q6)
df = pd.read_csv('/content/employee_productivity_dataset.csv')

# Redo Q1-Q6 preprocessing steps if starting fresh (or assume df is already processed)
# For simplicity, assuming df is already in the state after Q6 for pipeline definition

# Define numerical and categorical columns
numeric_features = ['Age', 'Salary', 'Hours_Worked_Per_Week',
                   'Performance_Score', 'Projects_Completed']

categorical_features = ['Gender', 'Department', 'Work_Mode', 'Location']

# Create preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),           # Standardization for numerical features
        ('cat', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'),
         categorical_features)                                 # One-Hot Encoding for categorical features
    ])

# Build the full pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

print("Pipeline built successfully!")
print(pipeline)
```

**Question 8: Apply Pipeline**

Apply the pipeline on the dataset.

Display:

 - Transformed dataset

 - Shape of final dataset

 **Answer:**

```python
# Apply the pipeline
X_transformed = pipeline.fit_transform(df)

# Convert back to DataFrame for better readability
feature_names = (
    numeric_features +
    list(pipeline.named_steps['preprocessor']
         .named_transformers_['cat']
         .get_feature_names_out(categorical_features))
)

transformed_df = pd.DataFrame(X_transformed, columns=feature_names)

print("Transformed Dataset (First 5 rows):")
print(transformed_df.head())

print("\nShape of final transformed dataset:", transformed_df.shape)
print("\nColumns in final dataset:", transformed_df.columns.tolist())
```

**Question 9: Why is scaling inportant in Python ?**

**Answer:** Scaling is crucial because many machine learning algorithms (e.g., SVM, KNN, Neural Networks, PCA) are distance-based or rely on gradients. Features with larger ranges (like Salary in thousands vs. Age in tens) can dominate the model, leading to biased results or slow convergence. Scaling puts all features on a similar scale, improving model performance and training stability.


**Question 10: Why do we convert categorical data into numerical from ?**

**Answer:** Machine learning models (except tree-based like Random Forest/XGBoost) work only with numerical inputs. Converting categories (via Label Encoding or One-Hot Encoding) allows the algorithm to process and find patterns in the data mathematically.

 - Label Encoding: Good for ordinal data.
 - One-Hot Encoding: Preferred for nominal data to avoid implying false order.
